```python
import os
import numpy as np
```

```python
def load_best_checkpoint(checkpoint_dir):
  ckpts = os.listdir(checkpoint_dir)
  if ckpts:
    ckpts.sort(key=lambda x:float(x.split('-')[-1][:-5]))
    return ckpts[0]
  else:
    return None
```

```python
def normalize(X): return X/np.sqrt(np.sum(X**2, axis=-1, keepdims=True)+1e-6)

def cosine_similarity(a,b,normalized=True):
  def dot(A,B): return (sum(a*b for a,b in zip(A,B)))

  if normalized: return dot(a,b)
  else: return dot(a,b)/((dot(a,a)**0.5)*(dot(b,b)**0.5))
```
